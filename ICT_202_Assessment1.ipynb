```python
import pandas as pd
import io
import numpy as np

# Added dataset to Github to prevent constant uploading issues.
url ='https://raw.githubusercontent.com/Devarthran/ICT202_Assign_1/main/dataset_information.csv'

df = pd.read_csv(url)
df # Uncomment to print dataset
```

# Data Preprocessing
## Steps taken
1. Encode class labels
2. Feature Selection using Recursive Feature Elimination (RFE)

```python
# Encoding class labels
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
# Applies numerical labels to the Class feature
df['Class'] = class_le.fit_transform(df['Class'].values)
# Sets all High_risk rows to 1 and Low_risk to 0
df['Class'] = df['Class'].apply(lambda x: 0 if x==1 else 1 if x==0 else x)

# Simplified for consideration, replaces 3 lines with 1 line of code.
# Replaces Class labels with numerical representation
#df['Class'] = df['Class'].replace({'High_risk':1,'Low_risk':0,'Nan':0})

# Uncomment to print df
df 
```

```python
# Feature Selection using RFE
# Seperating features from target class
x = df.drop('Class', axis=1)
y = df['Class']

from sklearn.model_selection import train_test_split

# Creating train/test split at 70/30 with a set random state for reproducable results.
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=0)
x_train.shape, x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier_rf = RandomForestClassifier(random_state=0, n_jobs=-1, max_depth=5,
                                       n_estimators=100, oob_score=True)

```

```python
%%time
# Cell magic to test performance
classifier_rf.fit(x_train, y_train)

# checking the oob score
classifier_rf.oob_score_
```
